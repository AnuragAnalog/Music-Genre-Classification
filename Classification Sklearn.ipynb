```python
# Required modules

import numpy as np
import pandas as pd

from matplotlib import pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression, SGDClassifier
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier
from sklearn.model_selection import GridSearchCV, RandomizedSearchCV
from sklearn.metrics import accuracy_score, classification_report
```

```python
# Load the 30 sec data features

features30 = pd.read_csv("./Data/features_30_sec.csv")
features30.head()
```

```python
# Load the 3 sec data features

features3 = pd.read_csv("./Data/features_3_sec.csv")
features3.head()
```

```python
# Seperating features and labels

X = features3.iloc[:, 2:-1]
y = features3["label"]
```

```python
# Split the data into features and target

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=88)
```

```python
# Encode the target

label_encoder = LabelEncoder()
y_train = label_encoder.fit_transform(y_train)
y_test = label_encoder.transform(y_test)
```

```python
# Standardize the data

scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

## Training the model

### Logistic Regression

```python
# Train the model

logreg = LogisticRegression()
logreg.fit(X_train, y_train)
```

```python
# Test the model

y_pred = logreg.predict(X_test)

# Evaluate the model

accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy}")
cr = classification_report(y_test, y_pred)
print(cr)
```

### Decision Tree

```python
# Train the model

dt = DecisionTreeClassifier()
dt.fit(X_train, y_train)
```

```python
# Test the model

y_pred = dt.predict(X_test)

# Evaluate the model

accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy}")
cr = classification_report(y_test, y_pred)
print(cr)
```

### KNN

```python
# Train the model

knn = KNeighborsClassifier()
knn.fit(X_train, y_train)
```

```python
# Test the model

y_pred = knn.predict(X_test)

# Evaluate the model

accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy}")
cr = classification_report(y_test, y_pred)
print(cr)
```

### AdaBoost

```python
# Train the model

ada = AdaBoostClassifier()
ada.fit(X_train, y_train)
```

```python
# Test the model

y_pred = ada.predict(X_test)

# Evaluate the model

accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy}")
cr = classification_report(y_test, y_pred)
print(cr)
```

### Random Forest Classifier

```python
# Train the model

model = RandomForestClassifier()
model.fit(X_train, y_train)
```

```python
# Test the model

y_pred = model.predict(X_test)

# Evaluate the model

accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy}")
cr = classification_report(y_test, y_pred)
print(cr)
```

```python
# Optimize the model using GridSearchCV

param_grid = {
    'n_estimators': [50, 100, 150, 200, 250],
    'max_depth': [5, 10, 15, None],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}

grid_search = GridSearchCV(estimator=model, param_grid=param_grid, cv=5, n_jobs=4, verbose=2)
grid_search.fit(X_train, y_train)

print(grid_search.best_params_)
print(grid_search.best_score_)
```
